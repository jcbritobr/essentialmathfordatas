```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 4, 5], [1, 4, 2, 4])
```

```python
np.random.seed(19680801)
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize= (5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data = data)
ax.set_xlabel("entry a")
ax.set_ylabel("entry b")
```

```python
x = np.linspace(-5, 2, 100)
y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2+ 10*x
y3 = 6*x + 10

fig, ax = plt.subplots()
ax.plot(x, y1, color = "blue", label = "y(x)")
ax.plot(x, y2, color = "red", label = "y'(x)")
ax.plot(x, y3, color = "green", label = "y''(x)")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```

```python
fig = plt.figure(figsize=(8, 2.5), facecolor="#616161")
left, bottom, width, height = .1, .1, .8, .8
ax = fig.add_axes((left, bottom, width, height), facecolor="#717171")
x = np.linspace(-2, 2, 1000)
y1 = np.cos(40 * x)
y2 = np.exp(-x**2)
ax.plot(x, y1 * y2, 'r')
ax.plot(x, y2, 'g')
ax.plot(x, -y2, 'g')
ax.set_ylabel("y")
ax.set_xlabel("x")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

def gensignal(time, frequency=3, sr=44100, amplitude=.5):
    T = np.linspace(0, time, sr)
    signal = amplitude * np.sin(2 * np.pi * frequency * T)
    return signal

time = 1
x = np.linspace(0, time, 44100)
signal1 = gensignal(time)
signal2 = gensignal(time, frequency=30, amplitude=1)
signal3 = gensignal(time, frequency=10, amplitude=.7)
signal4 = gensignal(time, frequency=60, amplitude=.5)

axes[0, 0].plot(x, signal1)
axes[0, 0].set_ylabel("power")

axes[0, 1].plot(x, signal2, color="blue")
axes[1, 0].plot(x, signal3, color="red")
axes[1, 1].plot(x, signal4, color="orange")
axes[1, 1].set_xlabel("time")

```
